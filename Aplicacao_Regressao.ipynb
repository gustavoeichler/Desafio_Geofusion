# Aplicar o modelo aos dados de SP

```python
import pandas as pd
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import plotly.express as px
import os
out = os.path.basename('C:/Users/guga_/Desafio Geofusion/Dados')

def Salva_DB(Caminho,Filename,Arquivo):
    open_file = open(f'{Caminho}/{Filename}', "wb")
    pickle.dump(Arquivo, open_file)
    open_file.close()

def Abre_DB(Caminho,Filename):
    with open(f'{Caminho}/{Filename}','rb') as f:
        DB = pickle.load(f)
    return DB
```

```python
import tensorflow as tf
new_model = tf.keras.models.load_model('Modelos/KeNN0.959')
new_model.summary()

```

```python
Base_SP = Abre_DB(f'{out}/Teste','Base_SP.pkl')
pd.DataFrame(Base_SP)
```

```python
faturamento = pd.DataFrame(new_model.predict(Base_SP))
faturamento

Salva_DB(out,'Faturamento_SP.pkl',faturamento)
```

```python
Base_Final = Abre_DB(out,'Base_SP.pkl')
Base_Final['faturamento']=np.array(faturamento)
```
